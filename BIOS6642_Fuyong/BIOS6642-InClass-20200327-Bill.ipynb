```python
#Ann's example, with diagnostic code
def del_dups_verbose(my_string):
    n = len(my_string)
    print("The initial length of the string is",n)
    
    #First base case: for strings where there is only one character left when adjacent duplicates are removed
    if (n<2):
        print("Offramp: n=",n,"(less than 2)")
        #This ends recursion and returns the string ("offramp")
        return my_string
    
    #We define a j to test the character after i in iteration
    #We need j instead of indexing my_string[i+1] because we will get an error if we index a character that isn't there.
    #Therefore, we need to check if the next character exists before indexing: see next comment.
    j=1
    
    for i in range(n):
        print("Beginning iteration. n=",n)
        print("Testing index",i)
        #To avoid an error, we must not index if j>=n (length of string).
        #This also becomes a second base case because we are done with recursion 
        #if there are no more adjacent duplicates to remove after iterating through the whole sting.
        if j >= n:
            print("Offramp: J is greater than n. The end of the string has been reached.")
            print("Returning my_string:",my_string)
            #"Offramp" 2
            return my_string

        #Recursive call: if we find an adjacent duplicate, we want to remove it and then iterate through the new string again.
        if my_string[j] == my_string[i]:
            print("A duplicate has been found at positions",i,"and",j)
            print("The duplicate characters are",my_string[i],"and",my_string[j])
            #This perfroms the removal using slicing and string concatenation
            my_string = my_string[:i] + my_string[j+1:]
            print("New string after deletion:", my_string)
            print("Beginning recursion")
            #After deletion, we begin recursion (or we begin more recursion)
            return del_dups_verbose(my_string)

        #Part of the for loop: with each iteration, we increase j to keep testing the next character
        print("Incrementing j by one.")
        j +=1
        print("j=",j)
    
    #"Offramp" 3: for black swan examples where the first two base cases are not tripped (I haven't found an example yet)
    print("The loop has ended. Returning my_string:",my_string)
    print("Other offramp (3): all for loops ended")
    return my_string
```

```python
my_string = input("Please enter a string with duplicate and adjacent letters. Can be lowercase or uppercase.")
my_string = my_string.lower()
my_string = list(my_string)
final = del_dups_verbose(my_string)

print(final)
#If the object to print is a list or tuple, the asterisk unpacks the items in the list or tuple and prints each individually
print(*final)
#This prints with no spaces between the characters, to make output cleaner
print(*final,sep='')
```

```python
#Removal using del. This requires a list as input.
def del_dups_2_verbose(my_string):
    n = len(my_string)
    print("The initial length of the string is",n)
    
    #First base case
    if (n<2):
        print("Offramp: n=",n,"(less than 2)")
        return my_string
    
    #Define a j to test the character after i in iteration
    j=1
    
    for i in range(n):
        print("Beginning iteration. n=",n)
        print("Testing index",i)
        #Second base case: looping through the whole string without finding duplicates
        if j >= n:
            print("Offramp: J is greater than n. The end of the string has been reached.")
            print("Returning my_string:",my_string)
            return my_string

        #Recursive call: duplicate discovered and removed
        if my_string[j] == my_string[i]:
            print("A duplicate has been found at positions",i,"and",j)
            print("The duplicate characters are",my_string[i],"and",my_string[j])
            #Deletion is achieved through the use of del
            #This will delete characters i-j 
            del my_string[i:j+1]
            print("New string after deletion:", my_string)
            print("Beginning recursion")
            #After deletion, we begin recursion (or we begin more recursion)
            return del_dups_2_verbose(my_string)

        #Part of the for loop: with each iteration, we increase j to keep testing the next character
        print("Incrementing j by one.")
        j +=1
        print("j=",j)
    
    #"Offramp" 3: for black swan examples where the first two base cases are not tripped (I haven't found an example yet)
    print("The loop has ended. Returning my_string:",my_string)
    print("Other offramp (3): all for loops ended")
    return my_string
```

```python
my_string = input("Please enter a string with duplicate and adjacent letters. Can be lowercase or uppercase.")
my_string = my_string.lower()
my_string = list(my_string)
final = del_dups_2_verbose(my_string)

#If the object to print is a list or tuple, the asterisk unpacks the items in the list or tuple and prints each individually
print(*final)
#This prints with no spaces between the characters, to make output cleaner
print(*final,sep='')
```
